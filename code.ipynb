```python
import requests

download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
target_csv_path = "nba_all_elo.csv"

response = requests.get(download_url)
response.raise_for_status()    # Check that the request was successful
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")


import pandas as pd

data = pd.read_csv('nba_all_elo.csv') 
data.head() #check the top part of data

l=[] 
dict_1={}
team=[]
for v in range(len(data)):
    l.append(str(data["fran_id"][v])) 
a= set(l)                               #listing all unique franchises
for u in range(len(data)):
    team.append(str(data["game_result"][u]))                #listing corresponding game results

for i in a:                            #counting the number of wins per team
        count=0   
        for x in range(len(data)):
            if i==str(data["fran_id"][x]) and team[x]=='W':              #converting obj 'game_result' to string type
                count=count+1
                
        print("\n the {} have won a total of {} games \n".format(i,count))           #listing of total wins of each team
        dict_1[i]=count    

greatest=0
val_list= list(dict_1.values())
key_list= list(dict_1.keys())
for r in range(len(val_list)):
    if greatest < int(val_list[r]):  #determining the most wins
        greatest=val_list[r]
for q in range(len(key_list)):           #determining the index of team with greatest wins
    if greatest == val_list[q]:
               index=q
print("\n the {} is the most succesfull team with {} wins \n".format(key_list[index],greatest))                
```
